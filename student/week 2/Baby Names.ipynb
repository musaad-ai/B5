```python
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv"
baby = pd.read_csv(url, sep=',')
```

```python
baby.head(10)
```

```python
baby.drop(columns=['Unnamed: 0', 'Id'])
```

```python
baby.groupby('Year').sum()['Count'].idxmax()
```

```python
baby["Gender"].value_counts()
```

```python
names = baby.groupby("Name")["Count"].sum()

names.sort_values(ascending=False).head()
```

```python
len(names)
```

```python
most = names.idxmax()
most
```

```python
lest = names.min()

lestt = (names == lest).sum()

lestt
```
