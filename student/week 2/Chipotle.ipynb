```python
dollarizer = lambda x: float(x[1:-1])
chip.item_price = chip.item_price.apply(dollarizer)
```

```python
revenue = (chip.quantity * chip.item_price).sum()
print('The revenue is '+str(revenue)+' dollars')
```
