# Exercise: Linear Regression on Dummy Dataset

You will **build a linear regression model** that fits the data: use the data to **train** a model, and the model will learn a line that best predicts $y$ from $X$.

### What the data is

The notebook uses **synthetic data** (fake data for practice):

- **`make_regression`** from scikit-learn creates:
  - **`X`**: 200 samples, 1 feature (one column of numbers).
  - **`y`**: 200 target values (what we want to predict).
- **`noise=20`**: adds randomness so the points don't lie exactly on a line.
- **`random_state=42`**: fixes the random numbers so you get the same data every run.

So you have 200 points in 2D: one input $x_1$ and one output $y$.

Let's make up some random regression data:

```python
from sklearn.datasets import make_regression

X, y = make_regression(
    n_samples=200,
    n_features=1,
    noise=20,
    random_state=42 # Set the seed for reproducibility
)
```

### Visualizing the data

The next cell plots the data:

- **`plt.plot(X, y, "b.")`**: scatter plot of $X$ vs $y$ (blue dots).
- **`xlabel` / `ylabel`**: labels for the horizontal axis ($x_1$) and vertical axis ($y$).
- **`plt.grid()`**: adds a grid to read values more easily.

You should see a **roughly linear** cloud of points: as $x_1$ increases, $y$ tends to increase (or decrease) in a line-like way. That's why linear regression is appropriate.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.plot(X, y, "b.")
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.grid()
plt.show()
```
